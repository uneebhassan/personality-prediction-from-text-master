```python
import requests
from selenium.webdriver import (Chrome, Firefox)
import pymongo
import datetime
from bs4 import BeautifulSoup
import pandas as pd
```

```python
browser = Chrome()
```

```python
url = 'https://twitter.com/search?q=%23dbfz&src=typd'
```

```python
browser.get(url)
```

```python
html = browser.page_source
```

```python
mc = pymongo.MongoClient()
```

```python
db = mc['tweets']
```

```python
dbfz_tweets = db['dbfz']
```

```python
dbfz_tweets.remove({})
```

```python
dbfz_tweets.insert_one(
    {
        'url': url,
        'datetime': datetime.datetime.now(),
        'html': html
    }
)
```

```python
dbfz_tweets.find_one()
```

```python
list(dbfz_tweets.find({'url': {'$exists': True}}))[0]
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
tweets = soup.select('div.tweet')
```

```python
def parse_tweet(tweet):
    username = tweet.select_one('span.username').text
    fullname = tweet.select_one('strong.fullname').text
    text = tweet.select_one('p.tweet-text').text
    tweet_time = tweet.select_one('span._timestamp').attrs['data-time']
    time = pd.Timestamp.fromtimestamp(int(tweet_time))
    tweet_dict = {
        'username': username,
        'fullname': fullname,
        'text': text,
        'tweet_timestamp': time
    }
    return tweet_dict
```

```python
parsed_tweet = [parse_tweet(tweet) for tweet in tweets]
pd.DataFrame(parsed_tweet)
```
